```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
```

```python
# Extract CSV into DataFrame
file1 = "Input\Jan_2022.csv"
file2 = "Input\Feb_2022.csv"
file3 = "Input\Mar_2022.csv"
file4 = "Input\April_2022.csv"
file5 = "Input\May_2022.csv"
file6 = "Input\June_2022.csv"
file7 = "Input\July_2022.csv"
file8 = "Input\August_2022.csv"
file9 = "Input\September_2022.csv"
file10 = "Input\October_2022.csv"
file11 = "Input\\Nov_2022.csv"
file12 = "Input\December_2022.csv"
```

```python
#Read CSV
jan_trip = pd.read_csv(file1, encoding="UTF -8")
feb_trip = pd.read_csv(file2, encoding="UTF -8" )
mar_trip = pd.read_csv(file3, encoding="UTF -8" )
apr_trip = pd.read_csv(file4, encoding="UTF -8" )
may_trip = pd.read_csv(file5, encoding="UTF -8" )
june_trip = pd.read_csv(file6, encoding="UTF -8" )
july_trip = pd.read_csv(file7, encoding="UTF -8" )
Aug_trip = pd.read_csv(file8, encoding="UTF -8" )
Sep_trip = pd.read_csv(file9, encoding="UTF -8" )
Oct_trip = pd.read_csv(file10, encoding="UTF -8" )
Nov_trip = pd.read_csv(file11, encoding="UTF -8" )
Dec_trip = pd.read_csv(file12, encoding="UTF -8" )
```

```python
# merge tables
citibike_df = [jan_trip, feb_trip,mar_trip,apr_trip,may_trip,june_trip,july_trip,Aug_trip,Sep_trip,Oct_trip,Nov_trip,Dec_trip]
```

```python
result = pd.concat(citibike_df)
```

```python
result.to_csv("Input\Trip_History.csv")
```
